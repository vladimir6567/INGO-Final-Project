```python
X = data[data.columns.values[2:-2:]].values
y = data[data.columns.values[-1::]].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
```

```python
X = imputer.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
comp = pca.fit_transform(X)

comp_0 = np.array([x for i, x in enumerate(comp) if y[i] == 0])
comp_1 = np.array([x for i, x in enumerate(comp) if y[i] == 1])

plt.scatter(comp_0[:, 0], comp_0[:, 1])
plt.scatter(comp_1[:, 0], comp_1[:, 1])
```

```python
from imblearn.under_sampling import ClusterCentroids
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4).fit(X)
rus = RandomUnderSampler(random_state=42,sampling_strategy=0.25)
cc = ClusterCentroids(random_state=1, sampling_strategy=1.0, estimator=kmeans)
#X_resampled, y_resampled = cc.fit_resample(X, y)
X_resampled, y_resampled = rus.fit_resample(X, y)
print(sorted(Counter(y_resampled).items()))
```

```python
from sklearn.decomposition import PCA

#pca = PCA(n_components=2)
comp = pca.transform(X_resampled)

comp_0 = np.array([x for i, x in enumerate(comp) if y_resampled[i] == 0])
comp_1 = np.array([x for i, x in enumerate(comp) if y_resampled[i] == 1])

plt.scatter(comp_0[:, 0], comp_0[:, 1])
plt.scatter(comp_1[:, 0], comp_1[:, 1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled,
                                                    test_size=0.2, random_state=42)
```

```python
pd.Series(y_test).hist()
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, pred)) 
```

```python
confusion_matrix(y_test, pred)
```

```python
importances = clf.feature_importances_
feature_names = data.columns[2:-2]


forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots()
forest_importances.plot.bar()
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
fn = feature_names[importances > 0.02].values
```

```python
fn
```

```python
X = data[fn].values
y = data[data.columns.values[-1::]].values

scaler = MinMaxScaler()
X = scaler.fit_transform(X)

imputer = KNNImputer(n_neighbors=3)

X = imputer.fit_transform(X)



kmeans = KMeans(n_clusters=4).fit(X)
rus = RandomUnderSampler(random_state=42, sampling_strategy=1.0)
cc = ClusterCentroids(random_state=1, sampling_strategy=1.0, estimator=kmeans)
#X_resampled, y_resampled = cc.fit_resample(X, y)
X_resampled, y_resampled = rus.fit_resample(X, y)
print(sorted(Counter(y_resampled).items()))
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled,
                                                    test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'n_estimators': [50, 75, 100, 150, 200], 'max_depth':[None, 3, 5, 10, 15, 20]}
```

```python
rdf = RandomForestClassifier()
clf = GridSearchCV(rdf, parameters, cv=5, verbose=2)

clf.fit(X_resampled, y_resampled)
```

```python
clf.best_score_, clf.best_params_
```

```python
rdf = RandomForestClassifier(**clf.best_params_)
rdf.fit(X_train, y_train)

pred = rdf.predict(X_test)

print(classification_report(y_test, pred)) 

confusion_matrix(y_test, pred)
```
